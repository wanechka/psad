* выборочное среднее
* несмещённую оценку стандартного отклонения
* возможность отклонения гипотезы о нормальности выборки по критерию Харке-Бэра, Шапиро-Уилка ($p \leq 0.05$);

Бинаризовать и посчитать для бинаризованной выборки:
*  сумму единиц
*  p-value для биномиального теста с двусторонней альтернативой ($p_0 = 0.5$)
*  значение статистики для критерия отношения правдоподобия


```python
import numpy as np
import pandas as pd
import scipy.stats as st
import seaborn as sns
import matplotlib.pyplot as plt

from statsmodels.stats.diagnostic import kstest_normal
from statsmodels.stats.weightstats import ttest_ind
from statsmodels.stats.weightstats import ztest
from statsmodels.stats.proportion import binom_test
from statsmodels.stats.proportion import proportion_confint
from statsmodels.stats.proportion import binom_test, proportions_ztest, proportions_chisquare
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Intelligent-Systems-Phystech/psad/master/hometask/sem2/tunev.i@phystech.edu.csv', header=None)
data_num = data.to_numpy().flatten()
```

Выборочное среднее

```python
print('mean = {:.2f}'.format(np.mean(data_num)))
```

Несмещённая оценка дисперсии

```python
print('unbiased STD estimation  {:.2f}'.format(np.std(data_num, ddof=1)))
```

Критерий Харке-Бэра

```python
JBS, pval = st.jarque_bera(data_num)
print('JB-statistics = {:.2f}, p-value = {:.2f}'.format(JBS, pval))

if pval <= 0.05:
  print("Reject")
else:
  print("Accept")
```

Критерий Шапиро-Уилка

```python
SHWS, pval = st.shapiro(data_num)
print('Shapiro-Wilk-statistics = {:.2f}, p-value = {:.2f}'.format(SHWS, pval))

if pval <= 0.05:
  print("Reject")
else:
  print("Accept")
```

Бинаризация выборки

```python
data_bin = np.ones(len(data_num))
data_bin[data_num < 0] = 0
```

Сумма единиц

```python
ones = int(np.sum(data_bin))
print('Ones number =', ones)
```

$p-value$ для точного биномиального теста

```python
k = ones
n = len(data_bin)
p0 = 0.5

pvalue = st.binom_test(k, n, p0, alternative='two-sided')
print('p-value = {:.2f}'.format(pvalue))

```
